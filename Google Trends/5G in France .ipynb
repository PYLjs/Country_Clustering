```python
#Importing libraries 
import pandas as pd
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
import plotly.express as px
```

```python
#Define Google search filters
kw_list = ['4G','5G']
pytrends = TrendReq(hl='FR',tz=360)
pytrends.build_payload(kw_list,cat=0,timeframe='2016-12-14 2020-12-31',geo='FR',gprop='')
```

```python
#Reading the data
df = pytrends.interest_over_time()
df
```

```python
#Cleaning the data
df = df.drop('isPartial',axis=1)
df
```

```python
#Visualize the data
plt.plot(df)
plt.xlabel('Date')
plt.ylabel('Search rate')
plt.title('4G and 5G Google searches in France through time')
plt.legend(df)
plt.show()
```

```python
#The 5G research trend is accelerating from the beginning of the year 2019, we will therefore focus on this period
focus = df.loc['2018-07-15':].plot()
plt.xlabel('Date')
plt.ylabel('Search rate')
plt.title('4G and 5G Google searches in France through time')
```

```python
#We are going to isolate three distinct periods during which we can observe a significant increase in google searches :

#-The month of February 2019 marks the start of the growth of the 5G trend in France. 
#Indeed, it was during this period that the first 5G deployment schedule was published. 
#In addition, the Mobile World Congress, which took place at the same time, was marked by the Orange communication (the leader of telecommunications operators in France) almost entirely devoted to 5G. 
#It is therefore natural that the media began to discuss 5G, wondering in particular about its impact on our daily lives.

#-Then, it was during the month of April 2020 that the four main French operators (Orange, SFR, Bouygues, Free) obtained official authorization to open a 5G network in France.

#-Finally, it is in October 2020 that the 5G trend increases a lot until it takes over the 4G trend. 
#Indeed, the 5G frequency auctions were conducted during this period. 
#After, the number of searches continues to increase and should not stop for several months.
```
